Sono attributi procedurali, cioè funzioni che possono essere usate solo nel contesto della classe in cui sono definiti<left>

```python
class Coordinate(object):
    def __init__(self,x,y):
        self.x = x
        self.y = y

        #metodi
    def distance(self, other):
        delta_x_sq = (self.x - other.x)**2
        delta_y_sq = (self.y - other.y)**2
        return (delta_x_sq + delta_y_sq)**0.5

c = Coordinate(3,4)
origin = Coordinate(0,0)

print(c.distance(origin))

print(Coordinate.distance(c, origin))

print (c)
```

<center>__str__<center>

E' l'analogo del toString() di java

```python
class Coordinate(object):
    def __init__(self,x,y):
        self.x = x
        self.y = y

        #metodi
    def distance(self, other):
        delta_x_sq = (self.x - other.x)**2
        delta_y_sq = (self.y - other.y)**2
        return (delta_x_sq + delta_y_sq)**0.5

    def __str__(self):
        return "<"+ str(self.x) + "," + str(self.y) + ">"

c = Coordinate(3,4)
print(c)
```

-- Esempio: creare un nuovo tipo per rappresentare un numero come FRAZIONE

La rappresentazione interna prevede due interi
• Numeratore
• Denominatore
§ Metodi
• Addizione, sottrazione
• Stampa della rappresentazione, conversione a float
• Inversione di una frazione

```python
#codice scritto da me😊

class Fraction(object):
    #variabili e init
    def __init__(self, num, den):
        self.num = num
        self.den = den

    #metodi
    def __add__(self, other):
        denominatore = self.den * other.den
        numeratore = self.num*other.den + self.den*other.num

        return Fraction(numeratore, denominatore)

    def __sub__(self, other):
        denominatore = self.den * other.den
        numeratore = self.num*other.den - self.den*other.num

        return Fraction(numeratore, denominatore)

    def __str__(self):
        return ("<" + str(self.num) + "/" + str(self.den) + ">")

    def __float__(self):
        return float(self.num)/float(self.den)

    def inverti(self):
        return Fraction(self.den, self.num)

#testing
f1 = Fraction(2,3)
f2 = Fraction(5,6)
print(f'f1: {f1}, f2: {f2}')
f3 = f1 + f2
f4 = f1-f2

print(f'somma: {f3}, \ndifferenza: {f4}')

print("Converto a float: ", float(f1))

print("Inverto: ", f1.inverti())
```

Metodi getter e setter

```python
class Animal(object):
    def __init__(self, age):
        self.age = age
        self.name = None

    #GETTER
    def get_age(self):
        return self.age
    
    def get_name(self):
        return self.name

    #SETTER
    def set_age(self, newage):
        self.age = newage

    def set_name(self, newname=""):
        self.name = newname

    def __str__(self):
        return "<animal: "+ str(self.name)+":"+str(self.age)+">"

a = Animal(22)
print(a)
a.set_name("Tizio")
print(a)    
```

<center> VARIABILI STATICHE (DI CLASSE)

```python
class Rabbit (Animal):
    tag = 1
    def __init__(self, age, parent1=None, parent2=None):
        Animal.__init__(self, age)
        self.parent1 = parent1
        self.parent2 = parent2

        self.rid = Rabbit.tag
        Rabbit.tag +=1

    def get_rid(self):
        return self.rid

r1 = Rabbit(11)
r2 = Rabbit(11)
r3 = Rabbit(20)
print(r1.get_rid(), r2.get_rid(), r3.get_rid())
r1 = Rabbit(11)
print(r1.get_rid())
```

Metodi statici -> metodi invocati sulla classe che non necessitano del parametro self

Devono essere preceduti dal decoratore @staticmethod

```python
class Rabbit (Animal):
    tag = 1
    def __init__(self, age, parent1=None, parent2=None):
        Animal.__init__(self, age)
        self.parent1 = parent1
        self.parent2 = parent2

        self.rid = Rabbit.tag
        Rabbit.tag +=1

    def get_rid(self):
        return self.rid

    @staticmethod
    def get_tag():
        return Rabbit.tag

r1 = Rabbit(11)
r2 = Rabbit(11)
r3 = Rabbit(20)
print(r1.get_rid(), r2.get_rid(), r3.get_rid())
r1 = Rabbit(11)
print(r1.get_rid())

Rabbit.get_tag()
```

<center>CLASSI ASTRATTE

Useremo il modulo _abc_ che fornisce l'infrastruttura per la definizione di _Abstract Base Class (ABC)_

```python
from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def doAction(self):
        pass

    def nonabstract_method(self):
        pass

class Human(Animal):
    def doAction(self):
        print("I can walk and run🥸")

class Snake(Animal):
    def doAction(self):
        print("I can crawl🐍")

class Dog(Animal):
    def doAction(self):
        print("I can bark🐶")

class Lion(Animal):
    def doAction(self):
        print("I can roar🦁")

#A=Animal() #--> non può farlo, essendo la classe astratta! [TypeError]
#A.doAction()

R = Human()
R.doAction()

K = Snake()
K.doAction()

R = Dog()
R.doAction()

K = Lion()
K.doAction()
```

<center>INTERFACCE FORMALI

In genere si prevede di sollevare un _NotImplementedError_

```python
from abc import ABC, abstractmethod

class MyInterface(ABC):  #Se qui NON specifico ABC, non mi darà l'errore!

    @abstractmethod
    def metodo1(self):
        raise NotImplementedError

    @abstractmethod
    def metodo2(self):
        raise NotImplementedError

    @abstractmethod
    def metodo3(self):
        raise NotImplementedError
```

```python
class MyInterfaceImpl(MyInterface):
    def metodo1(self):
        print("metodo1")

    def metodo2(self):
        print("metodo2")
    def metodo3(self):
        print("metodo3")

#c = MyInterface() #-> mi dà errore [TypeError]
d = MyInterfaceImpl()
d.metodo1()
d.metodo2()
d.metodo3()
```
