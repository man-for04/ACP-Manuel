<center>ECCEZIONI



Anche queste sono classi e ereditano dalla classe Exception

Come sono gestite? Con blocco try-catch, che diventa try-except

Nota che non specifico il tipo di eccezione, ma non è una buona pratica perchè non separo le gestioni!!

```python
try:
    a = int(input("Tell me one number:"))
    b = int(input("Tell me another number:"))
    print(a/b)
except:
    print("Bug in user input.")
```

```python
try:
    a = int(input("Tell me one number: "))
    b = int(input("Tell me another number: "))
    print("a/b = ", a/b)
    print("a+b = ", a+b)
except ValueError:
    print("Could not convert to a number.")
except ZeroDivisionError:
    print("Can't divide by zero")
except:
    print("Something went very wrong.")
```

Generalmente l'ultima è quella except (raccoglie tutto quello che non ho gestito)

else-> se quello che metto in try non solleva eccezioni

finally-> sempre eseguito, anche se non sollevo eccezioni

```python
try:
    x=int(input("Scrivi 1 per ok, 0 per eccezione: "))
    if x==0:
        raise Exception("descriptive string")
except Exception:
    print("errore!", Exception)
```

```python
def get_ratios(L1, L2):
    """ Assumes: L1 and L2 are lists of equal length of numbers
        Returns: a list containing L1[i]/L2[i] """
    ratios = []
    for index in range(len(L1)):
        try:
            ratios.append(L1[index]/L2[index])
        except ZeroDivisionError:
            ratios.append(float('nan')) #nan = not a number
        except:
            raise ValueError('get_ratios called with bad arg')
    return ratios

l1=[4,3,8,9]
l2=[7,9,0,1]
r = get_ratios(l1,l2)
print(r)
```

<center> ASSERZIONI

Voglio essere certo che certe assunzioni su una particolare condizioni siano come attese (vere)

Se non risulterà verificata, python consente attraverso _assert_ di sollevare un'eccezione di tipo _AssertionError_ (TERMINA PROCESSO)

```python
def avg(grades):
    assert len(grades) !=0 , 'no grades data'
    return sum(grades)/len(grades)

# ritorna le statistiche per soggetto
def get_stats(class_list):
    new_stats = []
    for elt in class_list:
        new_stats.append([elt[0], elt[1], avg(elt[1])])
    return new_stats


test_grades = [[['peter', 'parker'], [80.0, 70.0, 85.0]],
[['bruce', 'wayne'], [100.0, 80.0, 74.0]]]

test_grades = [[['peter', 'parker'], [80.0, 70.0, 85.0]], [['bruce', 'wayne'], []]]

print(get_stats(test_grades))
```
