dove nasce il problema delle liste? Se voglio memorizzare più informazioni su liste separate (sorta di database)

```python
name = ['Ana', 'John', 'Denise', 'Katy']
grade = ['B', 'A+', 'A', 'A']
course = [2.00, 6.0001, 20.002, 9.01]

#come ottengo una stampa completa?
def get_grade(student, name_list, grade_list, course_list):
    i=name_list.index(student)
    grade = grade_list[i]
    course = course_list[i]
    return (course, grade)

print(get_grade('John',name, grade, course))
```

```python
tupla = (1,2) #funzione
hash(tupla)
```
