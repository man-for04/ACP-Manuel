La sintassi *args permette per passare un _numero variabile di parametri_


E' un tipo che permette di inserire una lista di parametri (* NON è PUNTATORE)


```python
def fun(*args):
    print(type(args))
    for arg in args:
        print(arg)
fun("first", "second", "third")
```

```python
def fun(x, *ciao):
    print(type(ciao))
    print(x)
    for arg in ciao:
        print(arg)
fun(100, "first", "second", "third", 2)
```

KEYWORD ARGUMENTS:
Posso chiamare i parametri anche in un ordine diverso dalla firma:

```python
def fun(name, surname, country): #non metto ovviamente il tipo
    print(name, surname, country)
fun(surname="Rossi", country="Italy", name="Mario")
```

KEYWORD ARGUMENTS VARIABILI:
Posso gestire numero variabile con **

```python
def fun(**kwargs):
    print(type(kwargs))
    for key, value in kwargs.items():
        print(key, "=", value)
fun(day=8, month="March", year=2024)
```

```python
def fun(x, **kwargs):
    print(type(kwargs))
    print(x)
    for key, value in kwargs.items():
        print(key, "=", value)
fun("Date:", day=8, month="March", year=2024)
```

Posso usare funzione range per fare "packing" degli argomenti
*args viene automaticamente spacchettato in lista

```python
print(list(range(3, 6))) #normal call with separate arguments
# -> [3, 4, 5]
args = [3, 6]
print(list(range(*args))) #call with arguments unpacked from a list
# -> [3, 4, 5]
```
