with-> mantenere oggetto vivo (ottenere riferimento)

Oggetto avrà vita finchè mi trovo nel blocco istruzioni, l'ultima istruzione scatenerà in automatico il rilascio dell'oggetto che ho ottenuto alla prima istruzione

_with_ si occupa proprio di questa chiusura automatica


```python
with open('accounts.txt', mode='w') as accounts:
    accounts.write('100 Jones 24.98\n')
    accounts.write('200 Doe 345.67\n')
    accounts.write('300 White 0.00\n')
    accounts.write('400 Stone -42.16\n')
    accounts.write('500 Rich 224.62\n')
```

Posso iterare su un oggetto file grazie al costrutto _for_ :

viene così letta una riga alla volta

```python
for line in nomefile:  #sarà una stringa nel file
    print(line) 
```

```python
with open('accounts.txt', mode='r') as accounts:
    print(f'{"Account":<10}{"Name":<10}{"Balance":>10}') 
    for record in accounts:
        account, name, balance = record.split() #divide la stringa su account, name, balance (split senza parametri avviene sulla base di spazio)
        print(f'{account:<10}{name:<10}{balance:>10}')
```

--> la f specifica che (1.14)
I : nel print permettono una migliore formattazione (non obbligatorio ovviamente, se vuoi vedi)


Metodo readlines()-> leggere un intero file di testo, ritornando ogni riga del file come una stringa in una lista di stringhe 


```python
with open('accounts.txt', mode='r') as accounts:
    l=accounts.readlines()

print(l, type(l))
```

seek()-> permette di posizionare puntatore per lettura/scrittura


```python
with open('accounts.txt', mode='r') as accounts:
    accounts.seek(0)
    print(accounts.readline().strip())
```

.3f -> modalità formattazione per avere tot decimali, è un modo più "c like"

```python
import math
print(f'The value of pi is approximately {math.pi:.3f}.')
print(f'The value of pi is approximately {math.pi}.')
```

```python
table = {'Sjoerd':4127, 'Jack': 4098, 'Dcab': 7678}
for name, phone in table.items():
    print(f'{name:10} ==> {phone:10d}')
```
