Esercizio: Sviluppare una funzione
get_data(aTuple) che
- Estrae tutti gli interi da aTuple e li setta
come elementi in una nuova tupla.
- Estrae tutte le stringhe uniche da aTuple
e le setta come elementi in una nuova
tupla.
- Ritorna la tuple del minimo e del
massimo nella tupla di interi, e il numero
di stringhe unique.

```python
aTuple = ((3,"tre"),(2, "due"),(5,"due"))

def get_data(tupla):
    interi = ()
    stringhe = ()
    for t in tupla:
        interi = interi + (t[0],)
        if t[1] not in stringhe:
            stringhe = stringhe + (t[1], )
    minimo = min(interi)
    massimo = max(interi)
    return (len(stringhe), minimo, massimo)
    
t = get_data(aTuple)
print("Ci sono ", t[0], " stringhe uniche")
print("Tupla minima: ", t[1], ", tupla massima: ", t[2])
```

LISTE

Una sequenza ordinata di informazioni, accessibile tramite un indice
- Una lista è denotata da parentesi quadre, []
- Una lista contiene elementi di solito omogenei (i.e., tutti interi)
- può contenere valori di tipi misti (non comune)
- A differenza di una tupla, gli elementi di una lista possono cambiare
- Una lista è mutabile

```python
a_list = []
L = [2, 'a', 4, [1,2]]

print(len(L))
print(L[3])
```

Liste sono mutabili:

```python
L = [2,1,3]
L[1] = 99
print(L)
```

Posso iterare sugli elementi di una lista (es. somma):

```python
total = 0
for i in range(len(L)):
    total += L[i]

print(total)
```

```python
total = 0
for i in L:
    total += i

print(total)
```

LIST COMPREHENSION

E' un modo elegante per creare liste, l'espressione è:

_[expression for item in iterable]_

Dove:
- expression: Definisce il contenuto di ciascun elemento nella lista risultante
- item: Rappresenta l'elemento corrente su cui viene eseguita l'iterazione
- iterable: L'insieme o la sequenza su cui si itera
10

```python
fruits = ["apple", "banana", "cherry", "kiwi", "mango"]
newlist = []

for x in fruits:
    if 'a' in x:
        newlist.append(x) #oggetto mutabile

print(newlist)
print(type(newlist))
```

-- ALIAS: il cambiamento di una variabile impatta anche sull'altra

(essendo riferimenti allo stesso oggetto)

```python
a = 1
b = a
print("a: ", a)
print("b: ", b)

warm = ['red', 'yellow', 'orange']
hot = warm # creo alias
hot.append('pink')# modifico la lista mediante il suo alias
print(hot)
print(warm)
```

-- CLONARE UNA LISTA: creo una nuova lista e copio ogni elemento

(a differenza di alias, qui l'oggetto ORIGINARIO non viene influenzato)

```python
cool = ['blue', 'green', 'grey']
chill = cool[:]
chill.append('black') #modifico la COPIA
print(chill)
print(cool)
```

-- LISTE DI LISTE -> python permette di avere liste innestate (modificandole, si può incorrere in effetti collaterali)

```python
warm = ['yellow', 'orange']
hot = ['red']
brightcolors = [warm] #lista di lista 
brightcolors.append(hot)
print("lista di liste: ", brightcolors)
hot.append('pink')
print("\nlista interna modificata: ", hot)
print("\nlista di liste modificata: ",brightcolors)
```

```python
#Versione CORRETTA!
def remove_dups2(l1,l2):
    """voglio rimuovere gli elementi in comune da l1"""
    l1_bis = l1[:] #uso l1_bis per iterare
    for e in l1_bis:
        if e in l2:
            l1.remove(e)

L3 = [1,2,3,4]
L4 = [1,2,5,6]
remove_dups2(L3, L4)
print ("L1: ", L3, "\nL2: ", L4)
```
